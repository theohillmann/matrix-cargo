```python
import os
import pandas as pd
from joblib import dump


from src.features import Clustering
```

```python
notebook_path = os.getcwd()
root_path = os.path.dirname(os.path.dirname(os.path.dirname(notebook_path)))
root_path
```

```python
TRAIN_CSV_PATH = os.path.join(root_path, "CE263N Assignment 4/train.csv")
df = pd.read_csv(TRAIN_CSV_PATH)
df
```

```python
train_df = df.copy()
train_df.dropna(inplace=True)
clustering = Clustering()
train_df = clustering.create_columns(train_df)
train_df
```

```python
start_model, end_model = clustering.get_models()
start_model_path = os.path.join(root_path, os.path.join(notebook_path, "start_cluster_model.pkl"))
end_model_path = os.path.join(root_path, os.path.join(notebook_path, "end_cluster_model.pkl"))

dump(start_model, start_model_path)
dump(end_model, end_model_path)
```
